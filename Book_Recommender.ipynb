```python
ratings = pd.read_csv("BX-Book-Ratings.csv", encoding="latin-1", sep=";", on_bad_lines='skip')
ratings.head()
```

```python
ratings.columns
```

```python
ratings.rename(columns={'User-ID':'user_id',  'Book-Rating':'rating'}, inplace=True)
```

```python
#considering features on books , users, ratings
ratings.head(2)
```

```python
ratings['user_id'].value_counts().shape
```

```python
x=ratings['user_id'].value_counts()>200
```

```python
y=x[x].shape
```

```python
y=x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books= ratings.merge(books, on ='ISBN')
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'number of rating'}, inplace =True)
```

```python
number_rating
```

```python
final_rating= ratings_with_books.merge(number_rating, on ='title')
```

```python
final_rating
```

```python
final_rating=final_rating[final_rating['number of rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'], inplace= True)
```

```python
final_rating.shape
```

```python
book_pivot= final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace= True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse= csr_matrix(book_pivot)
```

```python
type(book_sparse)  #not going to consider 0 values
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model= NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.index==237
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
int(np.where(book_pivot.index == 'Along Came a Spider (Alex Cross Novels)')[0][0])
```

```python
def recommend_book(book_name):
    # Find indices where book_name matches
    indices = np.where(book_pivot.index == book_name)[0]
    
    if len(indices) == 0:
        print(f"Sorry, the book '{book_name}' was not found in the dataset.")
        return
    
    book_id = int(indices[0])
    distances, suggestions = model.kneighbors(
        book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6
    )
    
    print(f"Recommendations including the book itself:\n")
    for i in suggestions[0]:  # include the book itself
        print(book_pivot.index[i])
```

```python
recommend_book('Fine Things')
```

```python
recommend_book('good vibes good life')
```
